```python
# Importar librerias
import streamlit as st
from pathlib import Path
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Crgar y estandarizar  datos 

base_dir = Path.cwd()
csv_path = base_dir / 'data' / 'people-100.csv'
try:
    df = pd.read_csv(csv_path, sep=',', encoding='utf-8', header=0, index_col=0)
    print(' Archivo cargado correctamente.')
    df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_') 
except Exception as e:
    print(f" Error al cargar el archivo: {e}")
```

```python
# Grafico de barras En este caso tenemos pocos datos

grupo = df.groupby(['job_title', 'sex']).size().reset_index(name='count')
fig, ax = plt.subplots(figsize=(12, 6))
sns.barplot(data=grupo, x='job_title', y='count', hue='sex', ax=ax)
plt.xticks(rotation=45, ha='right')
plt.title('Distribucin de profesiones por sexo')
plt.tight_layout()

# st.pyplot(fig)
plt.show()
```

```python
#---------------------------------------------------------------------------------------------
    
    # Agrupar por job_title y contar ocurrencias
grupo = df['job_title'].value_counts().reset_index()
grupo.columns = ['job_title', 'count']

# Botn para mostrar grfico

fig, ax = plt.subplots(figsize=(12, 6))
sns.barplot(data=grupo, x='job_title', y='count', palette='Set2')
plt.xticks(rotation=45, ha='right')
plt.title('Frecuencia de profesiones')
plt.tight_layout()
# st.pyplot(fig)
plt.show()
```

```python




#---------------------------------------------------------
grupo = df['job_title'].value_counts().reset_index()
grupo.columns = ['job_title', 'count']


fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(grupo['job_title'], grupo['count'], color='teal', s=100, alpha=0.7)
ax.set_title("Dispersin de frecuencia por profesin")
ax.set_xlabel("Profesin")
ax.set_ylabel("Frecuencia")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
# st.pyplot(fig)
plt.show()
```

```python

# Conteo por sexo
conteo_sexos = df['sex'].value_counts()

# Grfico de barras
fig, ax = plt.subplots()
ax.bar(conteo_sexos.index, conteo_sexos.values, color=['skyblue', 'salmon'])
ax.set_title("Cantidad de personas por sexo")
ax.set_xlabel("Sexo")
ax.set_ylabel("Cantidad")
plt.tight_layout()
plt.show()
```
